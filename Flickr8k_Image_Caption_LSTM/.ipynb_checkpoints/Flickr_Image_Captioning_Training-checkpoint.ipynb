The following requires the Flickr8k dataset which contains images and 5 captions per image. The full dataset can be downloaded from AcademicTorrents.com.

In this notebook:
- The pre-trained CNN VGG16 is used to extract features from each image.
- The provided captions are cleaned and normalized to reduce the required vocabulary size, and then tokenized for modelling.
- A caption model is defined to combine image features with an LSTM to produce captions.
- Given the dataset size, the model is progressively trained and evaluated using BLEU score.
- An example image is then shown with a caption generated by the model

```python
# Import required libraries
from os import listdir
import numpy as np
import re
import json

from keras.applications.vgg16 import VGG16, preprocess_input
from keras.callbacks import ModelCheckpoint
from keras.layers import Add, Input, Dense, Dropout, Embedding, LSTM
from keras.models import Model, load_model
from keras.optimizers import Adam
from keras.preprocessing.image import load_img, img_to_array
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
from nltk.translate.bleu_score import corpus_bleu
```

```python
# Global bools for saving processed data to files, and loading saved data from files
save = True
load = False
```

## Image feature extraction using VGG16

```python
def extract_feat(folder_dir):
    """
    Utilizeses VGG16 classification model, pre-trained on imagenet dataset, to extract features of images, 
    located at given directory folder_dir.
    
    Args:
        folder_dir (str): A local folder directory containing all the images for which feature extraction is desired.
        
    Returns:
        dict: Contains numpy arrays of extracted features of each image located in folder_dir.
    """
    # Load pre-trained vgg16 model for new model
    vgg_model = VGG16(weights='imagenet', input_shape=(224, 224, 3))
    vgg_model.layers.pop()
    model = Model(inputs=vgg_model.input, outputs=vgg_model.layers[-1].output)
    
    features = {}
    # Extract features of each image in folder_dir
    for file in listdir(folder_dir):
        # Load image
        image_dir = folder_dir + '/' + file
        image = load_img(image_dir, target_size=(224,224))
        
        # Format image for model input
        image = img_to_array(image)
        image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
        image = preprocess_input(image)
        
        # Compute and store features
        image_feature = model.predict(image)
        features[file] = image_feature
        
    return features
```

```python
# Load all image features from file
if load: 
    features = np.load("caption_features.npy", allow_pickle=True)[0]
else:
    # Extract the features of the Flicker8K images
    folder_dir = "C:\Lantern_Data\Flickr8k\Flicker8k_Dataset"
    features = extract_feat(folder_dir)
    
    # Save all image features to file
    if save: np.save("caption_features.npy", np.array([features]))
```

## Preparing the captions

```python
def load_text(file_dir):
    """
    Opens the file located at file_dir, reads and returns the text.
    
    Args:
        file_dir (str): A local directory for the text file to be opened and read.
        
    Returns:
        str: The text contained in the file. 
    """
    file = open(file_dir, 'r')
    text = file.read()
    file.close()
    
    return text
```

```python
# Load all given captions from file
captions_dir = "C:\Lantern_Data\Flickr8k\Flickr8k.token.txt"
text = load_text(captions_dir)
```

```python
def format_captions(text):
    """
    Returns a dictionary mapping image filenames to the corresponding cleaned captions.
    
    Args:
        text (str): The text from the file containing the captions of images.
        
    Returns:
        dict: A mapping of each image file name to a set of possible captions.
    """
    captions = {}
    
    for line in text.split('\n'):
        # Split line and extract filename and caption
        tokens = line.split()
        image_id, image_cap = tokens[0].split('#')[0], tokens[1:]
        
        # Remove numbers, special characters and resulting single letters from captions
        image_cap = [re.sub(r'\W+', ' ', word) for word in image_cap]
        image_cap = [word.strip() for word in image_cap if (len(word)>1 and word.isalpha())]
        
        # Rejoin captions to single string and cast to lower
        image_cap = 'startseq ' + ' '.join(image_cap) + ' endseq'
        image_cap = image_cap.lower()
        
        # Add to dictionary
        if image_id not in captions:
            captions[image_id] = list()
        captions[image_id].append(image_cap)
    
    return captions
```

```python
# Load the cleaned captions from file
if load: 
    with open("Formatted_Captions.json", 'r') as file: captions = json.load(file)
else:
    # Extract and format captions from loaded file
    captions = format_captions(text)
    
    # Save the cleaned captions to new file
    save_as_txt = False
    if save:
        if save_as_txt:
            lines = []
            for image_id,image_cap_list in captions.items():
                for image_cap in image_cap_list:
                    lines.append(image_id + ' ' + image_cap)
            file = open("Formatted_Captions.txt", 'w')
            file.write('\n'.join(lines))
            file.close()
        else:
            with open("formatted_captions.json", 'w') as file:
                json.dump(captions, file)
```

```python
# Create a vocab of distinct words used in all the captions
vocab = set()
for key in captions.keys():
    [vocab.update(cap.split()) for cap in captions[key]]
print("Vocabulary size: {size} words".format(size=len(vocab)))
```

## Loading train and test sets

```python
def tokenize_captions(captions):
    """
    Returns a Keras Tokenizer object created uning captions.
    
    Args:
        captions (dict): Mapping of captions for each image.
        
    Returns:
        Tokenizer: Tokenized object of word in captions.
    """
    # Convert dict to list
    cap_list = []
    for key in captions.keys():
        [cap_list.append(i) for i in captions[key]]
    
    # Create Tokenizer object
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(cap_list)
    
    return tokenizer
```

```python
def max_cap_length(captions):
    """
    Returns the number of words contained in the longest sequence caption in captions.
    
    Args:
        captions (dict): Mapping of captions for each image.
    
    Returns:
        int: The largerst number of words per caption.
    """
    cap_list = []
    for key in captions.keys():
        [cap_list.append(i) for i in captions[key]]
    
    return max(len(cap.split()) for cap in cap_list)
```

```python
def sequence_data(tokenizer, captions, image_features, max_length):
    """
    Returns a series of Numpy arrays containing image and caption data useable by the caption 
    model to train/evaluate.
    
    Args:
        tokenizer (Tokenizer): Tokenizer object containing words from captions.
        captions (list): List of captions for the given image.
        image_features (array): The image features of the given image.
        max_length (int): The maximum number of words per caption.
        
    Returns:
        X1 (Numpy array): Ordered sequence of arrays of image features.
        X2 (Numpy array): Ordered sequence of lists containing preceding words of image caption.
        y (Numpy array): Ordered sequence of lists containing proceeding word of image caption.
    """
    X1,X2,y = [],[],[]
    vocab_size = len(tokenizer.index_word)
    
    # Iterate over captions for the image
    for cap in captions:
        index_seq = tokenizer.texts_to_sequences([cap])[0]
        # Split sequence into X1,X2,y pairs
        for i in range(len(index_seq)-1):
            input_seq, output_seq = index_seq[:i+1], index_seq[i+1]
            input_seq = pad_sequences([input_seq], maxlen=max_length)[0]
            output_seq = to_categorical([output_seq], num_classes=vocab_size+1)[0]
            
            X1.append(image_features)
            X2.append(input_seq)
            y.append(output_seq)
            
    return [[np.array(X1), np.array(X2)], np.array(y)]
```

```python
# Load training dataset
train = load_text("C:\Lantern_Data\Flickr8k\Flickr_8k.trainImages.txt").split('\n')
print("Training set size: {size} images".format(size=len(train)))
train_captions = {image_id:captions[image_id] for image_id in train}
train_features = {image_id:features[image_id] for image_id in train}
```

```python
# Define a maximum number of words for caption generation
max_len = max_cap_length(train_captions)
```

```python
# Retrieve tokens from training captions
tokens = tokenize_captions(train_captions)
print("Training set vocab size: {size} words".format(size=len(tokens.index_word)))
```

```python
# Load testing dataset
test = load_text("C:\Lantern_Data\Flickr8k\Flickr_8k.devImages.txt").split('\n')
print("Testing set size: {size} images".format(size=len(test)))
test_captions = {image_id:captions[image_id] for image_id in test}
test_features = {image_id:features[image_id] for image_id in test}
```

## Defining and training the captioning model

```python
def caption_model(vocab_size, max_length):
    """
    Returns a Functional Keras model that takes image features and caption sequences to generate the 
    next word of the captions.
    
    Args:
        vocab_size (int): The number of words in the vocabulary of the model.
        max_length (int): The maximum number of words per caption.
        
    Returns:
        Model: Keras model to generate image captions
    """
    # Features branch
    input_1 = Input(shape=(4096,))
    x_1 = Dropout(0.3)(input_1)
    x_1 = Dense(256, activation='relu')(x_1)
    
    # Sequence branch
    input_2 = Input(shape=(max_length,))
    x_2 = Embedding(vocab_size, 256, mask_zero=True)(input_2)
    x_2 = Dropout(0.3)(x_2)
    x_2 = LSTM(256)(x_2)
    
    # Merge branches
    y = Add()([x_1, x_2])
    y = Dense(256, activation='relu')(y)
    y = Dense(vocab_size, activation='softmax')(y)
    
    # Compile model
    model = Model(inputs=[input_1, input_2], output=y)
    model.compile(loss='categorical_crossentropy', 
                  optimizer=Adam())
    
    print(model.summary())
    
    return model
```

```python
# Create the training model
vocab_size=len(tokens.index_word)+1
model = caption_model(vocab_size, max_len)
```

```python
# Define training parameters
epochs = 5
train_steps = len(train_captions)
validate_steps = len(test_captions)
```

```python
# Iteratively train over epochs
for epoch_num in range(epochs):
    # Create batch generators
    train_generator = (sequence_data(tokens, image_cap_list, train_features[image_id][0], max_len) 
                       for i in iter(int, 1) for image_id, image_cap_list in train_captions.items())
    validate_generator = (sequence_data(tokens, image_cap_list, test_features[image_id][0], max_len) 
                          for i in iter(int, 1) for image_id, image_cap_list in test_captions.items())
    
    # Fit and save the model
    model.fit_generator(train_generator, steps_per_epoch=train_steps, 
                        validation_data=validate_generator, validation_steps=validate_steps)
    model.save('model-' + str(epoch_num) + '.h5')
```

## Evaluating the model

```python
def decode_index(integer, tokenizer):
    """
    Returns the word that corresponds to the given integer within the Tokenizer. Returns None if no word is found.
    
    Args:
        integer (int): An index that corresponds to a word contained in the Tokenizer object.
        tokenizer (Tokenizer): The tokenizer object used to encode and decode captions.
        
    Returns:
        str: A word corresponding to the given integer index in tokenizer.
    """
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    return None
```

```python
def generate_cap(model, tokenizer, image, max_len):
    """
    Returns a caption string generated using the provided model.
    
    Args:
        model (Model): Keras captioning model used to generate captions.
        tokenizer (Tokenizer): Tokenizer used to encode and decode words for modelling.
        image (array): The image features for the desired image caption.
        max_length (int): The maximum length of the caption.
        
    Returns:
        str: A caption generated by the model.
    """
    caption = 'startseq'
    
    for i in range(max_len):
        # Encode caption input to sequence
        sequence = tokenizer.texts_to_sequences([caption])[0]
        sequence = pad_sequences([sequence], maxlen=max_len)
        
        # Predict the next word in the sequence
        predictions = model.predict([image,sequence], verbose=0)
        
        # Extract index and convert to word
        word = decode_index(np.argmax(predictions), tokenizer)
        if word is None:
            break
        
        # Add word to the caption
        caption += ' ' + word
        
        if word == 'endseq':
            break
            
    return caption
```

```python
# evaluate the skill of the model
def evaluate_model(model, captions, images, tokenizer, max_length):
    """
    Uses the trained model to generate captions for each image and measures the BLEU score with 
    respect to the provided captions.
    
    Args:
        model (Model): Keras captioning model used to generate captions.
        captions (dict): Mapping of images to list of captions.
        images (dict): Mapping of images to feature arrays.
        tokenizer (Tokenizer): Tokenizer used to encode and decode words for modelling.
        max_length (int):
    """
    provided_cap, predicted_cap = [], []
    
    # Iterate over all provided images & captions
    for image_id, cap_list in captions.items():
        # Generate caption
        prediction = generate_cap(model, tokenizer, images[image_id], max_length)
        # Store provided and generated captions as list of words
        provided_cap.append([cap.split() for cap in cap_list])
        predicted_cap.append(prediction.split())
        
    # Compute and print the BLEU scores
    print('Cumulative 1-gram (BLEU-1) score: %f' % corpus_bleu(provided_cap, predicted_cap, weights=(1.0, 0, 0, 0)))
    print('Cumulative 2-gram (BLEU-2) score: %f' % corpus_bleu(provided_cap, predicted_cap, weights=(0.5, 0.5, 0, 0)))
    print('Cumulative 3-gram (BLEU-3) score: %f' % corpus_bleu(provided_cap, predicted_cap, weights=(0.33, 0.33, 0.33, 0)))
    print('Cumulative 4-gram (BLEU-4) score: %f' % corpus_bleu(provided_cap, predicted_cap, weights=(0.25, 0.25, 0.25, 0.25)))
```

```python
# Load evaluation dataset
eval_set = load_text("C:\Lantern_Data\Flickr8k\Flickr_8k.testImages.txt").split('\n')
print("Prediction set size: {size} images".format(size=len(eval_set)))
eval_captions = {image_id:captions[image_id] for image_id in eval_set}
eval_features = {image_id:features[image_id] for image_id in eval_set}
```

```python
# Load the caption generating model
if load:
    filename = 'model-4.h5'
    model = load_model(filename)

# Evaluate model on evalutation dataset
evaluate_model(model, eval_captions, eval_features, tokens, max_len)
```
